```python
wnl = WordNetLemmatizer()
```

```python
article1 = open('article1').read()
```

```python
article_words = word_tokenize(article1)
```

```python
stop_words = list(stopwords.words('english'))
stop_words.append(',')
stop_words.append('.')
words = [wnl.lemmatize(i.lower()) for i in article_words if i not in stop_words]
#from all words removing stopwords and then lemmatizing each word
```

```python
words_freq = nltk.FreqDist(words)
words_freq.most_common(5)
```

```python
most_common = [a[0]for a in words_freq.most_common(7)]
```

```python
#for each most common word finding the worst from the article which is most similar to it
max = 0
max_word = None
sim_list = []
for a in most_common[:5]:
    i = wordnet.synsets(a)[0]
    max_word = None
    max = 0
    for b in words:
        if len(wordnet.synsets(b)) != 0:
            j = wordnet.synsets(b)[0]
            if i != j :
                
                sim = i.wup_similarity(j)
        
                if sim != None and sim>=max:
                    max = sim
                    max_word = j               
    sim_list.append([i, max_word, max])
sim_list    
```

```python
for common_word in sim_list:
    G=nx.DiGraph()
    cname = common_word[0]
    G.add_node(cname.name())
    for hnym in cname.hypernym_paths()[0][::-1]:
        G.add_node(hnym.name())
        G.add_edge(cname.name(), hnym.name())
        cname = hnym
    pos = nx.spectral_layout(G)    
    nx.draw_networkx(G, pos, with_labels=True)   
    plt.show()
        
```
